```python
#利率计算
from math import exp,log
#1.有效年利率 EAR=(1+APR/n)n−1  名义年利率 10%，半年复利一次（一年 2 次）EAR=(1+0.1/2)2−1=1.052−1=10.25%
def effect(apr,nper):
    ear=(1+apr/nper)**nper-1
    return ear
print(round(effect(0.1, 2),5))
#2.不同复利频率的有效利率换算 EAR2​=(1+freq1​APR1​​)(freq1​/freq2​)−1 半年复利（2 次），APR=10%；按季度复利（4 次）的有效利率
#若要继续求apr，乘以freq2即可
def effcetiverates(apr1,fre1,fre2):
    ear2=(1+apr1/fre1)**(fre1/fre2)-1
    return  ear2
print(effcetiverates(0.1, 2, 4))
#3.名义利率 → 连续复利利率 rc​=freq×ln(1+rm) 名义利率 5%，半年复利rc​=2×ln(1+0.05)=2×0.04879≈0.09758
def rmtorc(rm,fre):
    rc=fre*log(1+rm)
    return rc
print(rmtorc(0.05, 2))
#4.连续复利终值计算 FV=PV×e**（rc​×t ）本金 100，连续复利利率 9.758%，投资 2 年 FV=100×e0.09758×2≈121.55
#与一般的复利终值FV=PV×(1+r)**n不一样
def fv_continuous(pv,rc,t):
    return pv*exp(rc*t)

```

```python
# 债券价格 PV(每期利率,总期数,每期利息,面值)
def bond_price(fv,ytm,couponrate,year,fre):
    rate=ytm/fre
    nper=year*fre
    pmt=fv*couponrate/fre
    return npf.pv(rate,nper,pmt,fv)
# 零息债券价格 现值 = 面值 / (1 + 利率)^ 期数
def zero_price(fv,rate,year):
    return npf.pv(rate,year,0,fv)
# 债券到期收益率YTM
def bond_ytm(pv,fv,couponrate,year,fre):
    nper=year*fre
    pmt=fv*couponrate/fre
    return npf.rate(nper,pmt,pv,fv)*fre
# 利率期限结构绘图
import matplotlib.pyplot as plt
time=[3/12,6/12,2,3,5,10,20]
rate=[0.47,0.6,1.18,1.53,2,2.53,3.12]
def plot_yield(time,rate):
    plt.plot(time,rate)
    plt.xlabel('time')
    plt.ylabel('rate')
    plt.title('term structure')
    plt.show()
print(plot_yield(time, rate))
```

```python
#麦考利久期计算  fv=1000，year=3，半年付息一次（fre=2），票面利率5%（crate），ytm=6%
def bond_price(fv,ytm,couponrate,year,fre):
    rate=ytm/fre
    nper=year*fre
    pmt=fv*couponrate/fre
    return npf.pv(rate,nper,pmt,fv)
def duration(year,fv,rate,couponrate,fre):
    nper=year*fre #（一共六次）
    coupon=fv*couponrate/fre
    b=bond_price(fv,rate,couponrate,year,fre)
    d=0
    for i in range(1,nper+1):
        t=i/fre
        cf=coupon if i<nper else coupon+fv
        pv_cf=npf.pv(rate/fre,i,0,cf)
        d+=t*(pv_cf/b)
    return round(d,4)
#麦考利久期修正=duration / (1+ytm / 复利次数)
def mod_d(dur,ytm,fre):
    return dur/(1+ytm/fre)
```
