```python
#CAPM模型
#y = α + β·x + ε
from scipy import stats
stock_ret=[0.065,0.0265,-0.0593,-0.001,0.0346]
mkt_ret=[0.055,-0.09,-0.041,0.045,0.022]
(beta,alpha,r,p,std_err)=stats.linregress(stock_ret,mkt_ret)
print(beta,alpha) #beta衡量个股收益率对市场收益率的敏感度（系统性风险）,alpha超额收益的估计值
print('r-square:',r**2) #衡量线性相关程度，r² 为判定系数（模型解释力）
print('p:',p) #用于检验回归系数是否显著（通常看 β 是否显著不为 0）
print(std_err) #估计系数的抽样误差
#可视化linear model graph
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)
alpha=1
beta=0.8
n=20
x=np.arange(n)
y=alpha+beta*x+np.random.rand(n)
z=alpha+beta*x
plt.plot(y,x)
plt.plot(z,x)
plt.show()
```

```python
# t 分布概率密度函数与绘图 stats.t.pdf(b, dfreedom)
#b是分位点，横坐标的值
#dfreedom自由度，通常为样本量减去估计的参数个数，自由度越小曲线越矮胖尾巴越厚，出现极端值的概率越大
#自由度越大，曲线越接近标准正态分布（钟形曲线
import scipy.stats as stats
a=range(-30,30)
dfreedom=30
x=[]
y=[]
for i in a:
    b=i/10
    x.append(b)
    y.append(stats.t.pdf(b,dfreedom))
plt.plot(x,y)
plt.title('t-density')
plt.show()
```

```python
#最小二乘法OLS
import statsmodels.api as sm
y=[1,2,3,4,2,3,4]
x=range(1,len(y)+1)
x=sm.add_constant(x)
results=sm.OLS(y,x).fit()
print(results.summary()) #系数估计值、标准误、t 统计量、p 值,模型整体 F 检验、R²、调
```
