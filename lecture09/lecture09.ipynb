# Portfolio Optimization using Sharpe Ratio

This project uses Python to calculate the optimal stock portfolio based on the Sharpe Ratio.

## Stocks Used

- IBM
- WMT
- C

## Libraries

```python
pandas
numpy
yfinance
scipy
```

## Main Steps

### 1. Download Stock Data

Use `yfinance` to download historical stock prices.

### 2. Calculate Annual Returns

Daily log returns are calculated first, then converted into annual returns.

### 3. Calculate Portfolio Variance

Portfolio risk is estimated using:

- correlation matrix
- standard deviation
- portfolio weights

### 4. Calculate Sharpe Ratio

### 5. Optimize Portfolio Weights

Use `scipy.optimize.fmin()` to find the portfolio weights with the highest Sharpe Ratio.

---

# Example Output

```python
Optimal weights:
IBM: 0.25
WMT: 0.60
C: 0.15

Optimal Sharpe Ratio:
1.12
```

````python
import pandas as pd
import numpy as np
import yfinance as yf
from scipy.optimize import fmin

tickers = ['IBM', 'WMT', 'C']

begdate = '2010-01-01'
enddate = '2012-12-31'

rf = 0.003          # risk-free rate
n = len(tickers)

# -----------------------------------
# Step 1: annual return
# -----------------------------------

def ret_annual(ticker, begdate, enddate):

    # download data
    df = yf.download(ticker, start=begdate, end=enddate)

    # daily log return
    df['logret'] = np.log(df['Close'] / df['Close'].shift(1))

    # extract year
    df['year'] = df.index.year

    # annual return
    retannual = np.exp(df['logret'].groupby(df['year']).sum()) - 1

    # convert to DataFrame
    retannual = pd.DataFrame(retannual)

    retannual.columns = ['ret_' + ticker]

    return retannual


# -----------------------------------
# Step 2: portfolio variance
# -----------------------------------

def portfolio_var(r, w):

    # correlation matrix
    cor = np.corrcoef(r.T)

    # standard deviation
    std_dev = np.std(r, axis=0)

    var = 0

    for i in range(n):
        for j in range(n):

            var += (
                w[i]
                * w[j]
                * std_dev[i]
                * std_dev[j]
                * cor[i, j]
            )

    return var


# -----------------------------------
# Step 3: Sharpe Ratio
# -----------------------------------

def sharperatio(r, w):

    # portfolio variance
    var = portfolio_var(r, w)

    # mean return of each stock
    mean_return = np.mean(r, axis=0)

    # portfolio return
    port_return = np.dot(w, mean_return)

    # Sharpe Ratio
    sr = (port_return - rf) / np.sqrt(var)

    return sr


# -----------------------------------
# Step 4:
# optimize n-1 weights
# -----------------------------------

def sharperatio_n_minus_1_stocks(w):

    # calculate final weight
    w2 = np.append(w, 1 - sum(w))

    # negative Sharpe Ratio
    # because fmin minimizes
    return -sharperatio(r, w2)


# -----------------------------------
# collect annual returns
# -----------------------------------

x2 = ret_annual(tickers[0], begdate, enddate)

for ticker in tickers[1:]:

    x_ = ret_annual(ticker, begdate, enddate)

    x2 = x2.merge(
        x_,
        left_index=True,
        right_index=True
    )

# convert to numpy array
r = np.array(x2)

# -----------------------------------
# equal-weight portfolio
# -----------------------------------

print('-----------------------------------')
print(f'Tickers used: {tickers}')
print('-----------------------------------')

equal_w = np.ones(n) / n

equal_sr = sharperatio(r, equal_w)

print('Sharpe Ratio with equal weights:')
print(equal_sr)

# -----------------------------------
# optimization
# -----------------------------------

# initial guess
w0 = np.ones(n - 1) / n

# optimization
w1 = fmin(sharperatio_n_minus_1_stocks, w0)

# final weights
optimal_w = np.append(w1, 1 - sum(w1))

# optimal sharpe ratio
optimal_sr = sharperatio(r, optimal_w)

# -----------------------------------
# results
# -----------------------------------

print('\nOptimal weights:')

for i in range(n):
    print(f'{tickers[i]}: {optimal_w[i]:.4f}')

print('\nOptimal Sharpe Ratio:')
print(optimal_sr)
````
